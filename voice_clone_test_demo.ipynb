```python
from gtts import gTTS

# Your German text
german_text = "Hallo! Wie geht es dir heute? Ich hoffe, du hast einen schönen Tag."

# Convert text to speech
tts = gTTS(text=german_text, lang='de')

# Save the audio file
tts.save("german_audio.mp3")

print("Audio file saved as 'german_audio.mp3'")
```

```python
from gtts import gTTS

bangla_text = "হ্যালো! তুমি কেমন আছো? আমি তোমার জন্য একটি অডিও তৈরি করেছি।"
tts = gTTS(text=bangla_text, lang='bn')
tts.save("bangla_audio.mp3")

print("Bangla audio saved as 'bangla_audio.mp3'")
```
